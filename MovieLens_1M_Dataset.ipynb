```python
unames = ["user_id","gender","age","occupation","zip"]
users = pd.read_table(r"C:\Users\Gurushanth S S\Downloads\ml-1m\ml-1m\users.dat", sep="::", header=None, names=unames, engine="python")
```

```python
rnames = ["user_id","movie_id","rating","timestamp"]
ratings = pd.read_table(r"C:\Users\Gurushanth S S\Downloads\ml-1m\ml-1m\ratings.dat", sep="::", header=None, names=rnames, engine="python")
```

```python
mnames = ["movie_id", "title","genres"]
movies = pd.read_table(r"C:\Users\Gurushanth S S\Downloads\ml-1m\ml-1m\movies.dat", sep="::", names=mnames, engine="python", encoding="latin1")
```

```python
movies.head()
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data
```

```python
data.iloc[0]
```

```python
#To get mean movie ratings for each film grouped by gender
```

```python
mean_ratings = data.pivot_table("rating", index="title", columns="gender", aggfunc="mean")
```

```python
mean_ratings.head()
```

```python
#filter down to movies that received at least 250 ratings (an arbitrary number)
```

```python
ratings_by_title = data.groupby("title").size()
```

```python
ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title>=250]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
```

```python
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values("F", ascending=False)
```

```python
top_female_ratings.head()
```

```python
#Measuring Rating Disagreement

#to find the movies that are most divisive between male and female viewers
```

```python
mean_ratings["diff"]=mean_ratings["M"]-mean_ratings["F"]
```

```python
sorted_by_diff= mean_ratings.sort_values("diff")
```

```python
sorted_by_diff.head()
```

```python
sorted_by_diff[::-1].head()
```

```python
rating_std_by_title = data.groupby("title")["rating"].std()
```

```python
rating_std_by_title = rating_std_by_title.loc[active_titles]
```

```python
rating_std_by_title.head()
```

```python
movies["genre"]=movies.pop("genres").str.split("|")
```

```python
#Now, we can merge all three tables together and group by genre:

```

```python
ratings_with_genre=pd.merge(pd.merge(movies_exploded,ratings),users)
```

```python
ratings_with_genre.iloc[0]
```

```python
genre_ratings = (ratings_with_genre.groupby(["genre","age"])["rating"].mean().unstack("age"))
```
